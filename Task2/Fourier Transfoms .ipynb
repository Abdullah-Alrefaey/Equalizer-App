##### Trying with Sinusoids 

```python
def sin(c, f, phi):
    """
    c: Coefficient of the Equations
    f : Frequency 
    phi: phase Shift 
    return : c*sin(2*pi*f*t+ phi)
    """
    t = np.arange(0, 1.0, 0.01)
    pi = np.pi
    x = c*np.sin(2*pi*f*t + phi)
    return x
```

```python
from scipy import fftpack
```

###### you can see that it detects the frequencies with which the signal is formed 

```python
x = sin(1, 5, 0) + sin(2, 10, 0) + sin(3, 20, 0)
x_ft = fftpack.fft(x) 
y = np.abs(x_ft) * 2 / len(x)
freqs = fftpack.fftfreq(x.size, d = 0.01)
plt.axes(xlim=(0, 20))

plt.plot(freqs, y)
plt.xlabel("Frequencies")
plt.ylabel("Magnitude of pLower")
plt.show()
```
