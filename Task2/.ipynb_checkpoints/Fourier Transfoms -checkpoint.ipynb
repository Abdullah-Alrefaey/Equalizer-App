### Testing the Fourier Transform

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy as sp 
%matplotlib inline 
```

```python
import pandas as pd 
```

##### Trying with Sinusoids 

```python
def sin(c, f, phi):
    """
    c: Coefficient of the Equations
    f : Frequency 
    phi: phase Shift 
    return : c*sin(2*pi*f*t+ phi)
    """
    t = np.arange(0, 1.0, 0.01)
    pi = np.pi
    x = c*np.sin(2*pi*f*t + phi)
    return x
```

```python
x = sin(1, -5, 0)
print(x)
```

```python
plt.plot(x)
plt.title("Sine Wave")
plt.show()
```

```python
from scipy import fftpack
```

```python
x_ft = fftpack.fft(x)
y = np.abs(x_ft)
freqs = fftpack.fftfreq(x.size, d=0.01)

plt.plot(freqs, y)
plt.xlabel("Frequencies")
plt.ylabel("Magnitude of pLower")

plt.show()
```

```python
x = sin(1, 5, 0.3*np.pi) + sin(1, 10, 0) + sin(1, 20, 0)
x_ft = fftpack.fft(x)
y = np.abs(x_ft)
freqs = fftpack.fftfreq(x.size, d = 0.01)
plt.axes(xlim=(-20, 20))

plt.plot(freqs, y)
plt.xlabel("Frequencies")
plt.ylabel("Magnitude of pLower")
plt.show()
```

###### you can see that it detects the frequencies with which the signal is formed
